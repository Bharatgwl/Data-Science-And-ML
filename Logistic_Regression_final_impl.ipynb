```python
def sigmoid(z):
    return 1 /( 1 + np.exp(-z))
```

```python



def gd(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    epochs = 4000
    for i in range(epochs):
        y_hat = sigmoid(np.dot(X, weights))
        weights = weights + lr * (np.dot(X.T, (y - y_hat)) / X.shape[0])
    return weights[1:], weights[0]
```

```python
coef_, intercept_ = gd(X, y)
```

```python
print(coef_, intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
x_input2 = np.linspace(-3,3,1000)
y_input2 = m*x_input2 + b
plt.figure(figsize=(10,6))
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.plot(x_input2,y_input2,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.legend(['Logistic Regression','Gradient Descent'])
plt.ylim(-3,2)
```
