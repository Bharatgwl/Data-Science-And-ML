```python
 import matplotlib.pyplot as plt 
 %matplotlib inline
 import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## plotting using matplotlib 
## plt scatter

plt.scatter(x,y, c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2d-graph')
```

```python
y = x*x
plt.plot(x,y, 'mh-',linestyle='dashed', linewidth=0.5, markersize=5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('x y axis')
```

```python
#creating subplots 

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'yo')
plt.subplot(2,2,4)
plt.plot(x,y,'m-o')
```

```python
x = np.arange(0, 4* np.pi,0.1)
y = np.sin(x)
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.title('sine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
y = np.cos(x)
plt.subplot(2,2,2)
plt.plot(x,y,'m--')
plt.title('cos wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

y = (np.cos(x)/np.sin(x))
plt.subplot(2,2,3)
plt.plot(x,y,'y--')
plt.title('cot wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

# bar-graph

```python
#bar-graph

x = np.arange(1,11,2)

y = np.arange(11,21,2)*x

plt.bar(x,y , color='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('bargraph')
```

```python
# histograms

a =  np.array([22,55,77,59,69,47,4,33,44])
plt.hist(a,bins=20,align='left')
```

```python
# Box-plots using matplotlib

data = [np.random.normal(0,std,100) for std in range(1,4)]

#rectangular boxplots
plt.boxplot(data,vert=True, patch_artist=True)
plt.show()
```

# Pie chart

```python
# Data to plots 
labels = 'python' , 'c++' ,'ruby','Java'
sizes = [300,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'skyblue']
explode = (0.1,0,0,0)

plt.pie(sizes,explode=explode , labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.show()
```
