## correlation with heatmap

###### correlation heatmap uses colored cells typically in a maonochormatic scale to show a 2d correlation matrix (table) between
###### two discrete dimensions or events types. it is very important in features Selection it will only used when the values of the 
###### data is integer or floating other wise it will not work because the categorical features are the object type they cant be in negative 
* when we are applying correlation our value will ranges between only -1 to +1
* because of the pearson coeffiecient correlation 

```python
df.corr(numeric_only=True)  #this signify that if the total bill is increasing then tip will also increases
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

# join plot 

> join plot help to study the relationship between numeric variables The central chart display their correlation it is usually
> a scatter plot , a hexbin plot a 2d histogram or a 2D density plot

 ### univariate 

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# pair plot 
> A pair plot is also known as a scatterplot in which one variable in the same data row is matched with another variables
> values lik these pairs are just elaboration on this showing all variable paired with all the other variables 

```python
## it is the combination of all plots of dependent value of integer and float only 
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

# distplot

> distplot help us to check the distribution of the coloumn features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
