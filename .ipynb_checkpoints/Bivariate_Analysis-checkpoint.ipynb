# 4. distplot

```python
sns.displot(titanic['age'],kde=true,color='darkred', bins=40)
```

```python
from matplotlib.pyplot import hist
from sympy import false


sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

# 8. Lineplot (NUM NUM)

```python
flight = sns.load_dataset('flights')
```

```python
flight
```

```python
newdf=flight.groupby('year')['passengers'].sum().reset_index()
```

```python
newdf
```

```python
sns.lineplot(x='year',y='passengers',data=newdf)
```

```python
pivot =flight.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(pivot)
```
